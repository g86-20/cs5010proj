```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn import preprocessing
```

### Reads in CSV files as DataFrames and normalizes the data:

```python
aapl = pd.read_csv('./AAPL_20200321-183443.csv', index_col=0)
```

```python
pfe = pd.read_csv('./PFE_20200321-184056.csv', index_col=0)
```

```python
stz = pd.read_csv('./STZ_20200321-184000.csv', index_col=0)
```

```python
cv = pd.read_csv('./coronavirus_google.csv', skiprows=2)
cv = cv.rename(columns={"Day": "Date"}) # so that it matches up with the stocks dataframes 
```

```python
def normalize(df, col):
    # normalizes a dataframe column and adds the normalized values to a new column
    min_max_scaler = preprocessing.MinMaxScaler()
    col_vals = df[[col]].values
    norm_col = min_max_scaler.fit_transform(col_vals)
    df['norm'] = norm_col
```

```python
normalize(aapl, 'Adj Close Price')
normalize(pfe, 'Adj Close Price')
normalize(stz, 'Adj Close Price')
```

```python
# changes the values that were '<1' to zero
cv.loc[cv['coronavirus: (United States)']=='<1',['coronavirus: (United States)']] = 0
```

```python
normalize(cv, 'coronavirus: (United States)')
```

```python
stocks = [aapl, pfe, stz]
```

```python
words = [cv]
```

```python
from datetime import datetime
aapl['Date'] = pd.to_datetime(aapl['Date'])
pfe['Date'] = pd.to_datetime(pfe['Date'])
stz['Date'] = pd.to_datetime(stz['Date'])
```

### Plotting:

```python
def create_plot(words, stocks):
    fig = go.Figure()
    for word in enumerate(words):
        fig.add_trace(go.Scatter(x=word[1]['Date'], y=word[1]['norm'],
                                    mode='lines',
                                    name='word '+str(word[0]+1))) # change later 
    for stock in enumerate(stocks):
        fig.add_trace(go.Scatter(x=stock[1]['Date'], y=stock[1]['norm'],
                                    mode='lines',
                                    name='stock '+str(stock[0]+1))) # change later 
    fig.update_layout(title='Inverted \'Coronavirus\' searches vs. Stock Prices',
                       xaxis_title='Date',
                       yaxis_title='Normalized Stock Price or Search Frequency')
    fig.show()
```

```python
create_plot(words, stocks)
```
